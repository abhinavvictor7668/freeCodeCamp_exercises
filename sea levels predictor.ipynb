```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

df = pd.read_csv('sea levels.csv')
```

```python
def draw_plot():
    
    plt.figure(figsize=(10, 6))
    plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'], label='Original Data', alpha=0.6)

    res1 = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
    years1 = pd.Series(range(1880, 2051))
    plt.plot(years1, res1.intercept + res1.slope * years1, 'red', label='Best fit: 1880-2050')

    df_recent = df[df['Year'] >= 2000]
    res2 = linregress(df_recent['Year'], df_recent['CSIRO Adjusted Sea Level'])
    years2 = pd.Series(range(2000, 2051))
    plt.plot(years2, res2.intercept + res2.slope * years2, 'green', label='Best fit: 2000-2050')

    plt.xlabel('Year')
    plt.ylabel('Sea Level (inches)')
    plt.title('Rise in Sea Level')
    plt.legend()
    plt.show()

draw_plot()
```
